# the effect of the P component 
The steering angle is proportional to the cross track error. If the cross track error of the vehicle is big, the steering angle will be big and the vehicle are more willing to turn towards the target trajectory, when the vehicle gets closer and closer to the target trajectory, the steering angle will become smaller and smaller. 

# the effect of the D component 
The effect of the D component is to avoid the overshoot. If the slope of the curve is large as the vehicle approaches the target trajectory, like the red line in the picture, the vehicle won't just go shooting, instead, it will counter steers. This will stop the vehicle from moving too quickly towards the target trajectory.

# the effect of the I component
The effect of the I component is to avoid systematic bias. If a zero steering angle no longer keeps the vehicle driving straight, the vehicle will experience a constant cross track error. Since the I component is going to sum all the cross track errors in the past, If the vehicle experience a constant cross track error, this term will be large, as a result the vehicle will have a large steering angle to make the vehicle move torward the target trajectory.

# Describe how the final hyperparameters were chosen.
I tune the hyperparameters manually. First, I set all three components to zero, then I increase the P component until the vehicle can stay in the track for a distance. Next, I start adding the D component until the oscillation is small enough. The vehicle can successfully drive a lap around the track without tuning the I component. There aren't a lot of difference with or without I component in this project.
